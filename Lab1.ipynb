```python
import pandas as pd
import seaborn as sns


# Загрузка данных
file_path = 'D:\\FIT\\4-2\\КСВД\\Lab1\\shopping_trends.csv'

data = pd.read_csv(file_path)

print(data.head())
```

```python
# Перевірка розміру таблиці
print("Розмір таблиці:", data.shape)

# Перевірка наявності пропусків
print("Пропуски в даних:\n", data.isnull().sum())

# Перевірка наявності дублікатів
print("Дублікати в даних:", data.duplicated().sum())

# Типи стовпців
print("Типи стовпців:\n", data.dtypes)

# Перейменування назв стовпців
data.rename(columns=lambda x: x.lower().replace(' ', '_'), inplace=True)

# Перевірка результатів перейменування
print("Нові назви стовпців:\n", data.columns)
```
