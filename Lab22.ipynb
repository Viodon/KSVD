```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dense
```

```python
# Задаємо кількість даних
N = 10000

# Генеруємо дані
data = np.cos(3 * np.linspace(0, 10, N)) + 0.1 * np.random.randn(N)

# Побудова графіка для візуалізації даних
plt.plot(data)
plt.xlabel('Часовий крок')
plt.ylabel('Значення функції')
plt.title('Функція з шумом')
plt.show()
```

```python
# Розмір даних
N = 10000

# Створення даних
data = np.sin(np.arange(N) / 5) + np.cos(3 * np.arange(N))

# Розмір вхідної послідовності
input_shape = (1, 1)

# Побудова моделі
model = Sequential()
model.add(Bidirectional(LSTM(64, activation='relu'), input_shape=input_shape))
model.add(Dense(1))

# Компіляція моделі
model.compile(optimizer='adam', loss='mse')

# Навчання моделі
history = model.fit(data[:-1].reshape(-1, 1, 1), data[1:].reshape(-1, 1), epochs=10, batch_size=32)

# Побудова графіка втрат на тренувальному наборі
plt.plot(history.history['loss'], label='Тренувальні втрати')
plt.xlabel('Епохи')
plt.ylabel('Втрати')
plt.title('Графік втрат')
plt.legend()
plt.show()
```
